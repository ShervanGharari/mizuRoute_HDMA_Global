# Creating the point shapefile from the streamflow nc files, read the CSV of seg_id and populate lat lon

```python
# read and match data to the 
# relating_df = pd.read_csv('../data/D09/D09_925.HDMA.big-rivers.asc')
relating_df = pd.read_csv('../data/D09/D09_925.HDMA.csv')
relating_df ['lat'] = None
relating_df ['lon'] = None

# read the dataset
ds = xr.open_dataset('../data/D09/coastal-stns-Vol-monthly.updated-May2019.mod.nc', decode_times=False)

# add lat, lon, id
df = pd.DataFrame()
df['lat'] = ds['lat'][:]
df['lon'] = ds['lon'][:]
df['id']  = ds['id'][:]

# adding lat lon to the relating_df
for index, row in relating_df.iterrows():
    gauge_id = row.gauge_id
    df_slice = df[df['id']==gauge_id]
    relating_df.lat.loc[index] = df_slice.lat.iloc[0]
    relating_df.lon.loc[index] = df_slice.lon.iloc[0]
    

# initializing EASYMORE object
esmr = easymore()
shp_points = esmr.make_shape_point(relating_df,
                                   'lon',
                                   'lat')
shp_points.plot()


# name of target river 
# target_rivers = ['Nelson','Mackenzie','St Lawrence','San Juan','Euphrates (Fira',\
#                  'Parana','Yenisey','Changjiang','Nile','Volta','Zambeze','Karun',\
#                  'Karkheh','Mississippi','Rhine']
target_rivers = ['Nelson','Mackenzie','Parana','Rhine']

relating_df = relating_df[relating_df['riv_name'].isin(target_rivers)]

print(relating_df)

dictionary = dict(zip(relating_df.riv_name, relating_df.route_id))

dictionary_rename = dict(zip(relating_df.gauge_id.astype(str), relating_df.riv_name))

print(dictionary)
print(dictionary_rename)
```

```python
# read the dataset
ds = xr.open_dataset('../data/D09/coastal-stns-Vol-monthly.updated-May2019.mod.nc', decode_times=False)

# FLOW
FLOW = np.array(ds['FLOW'][:])
COLUMNS = list(np.array(ds['id'][:]).astype(str))

# dataindex
dateindex = pd.date_range('1900-01','2018-12', freq= 'MS')+pd.DateOffset(days=14) # first year 190001, end 201812
dateindex = pd.date_range('1900-01','2018-12', freq= 'MS')+pd.DateOffset(days=0) # first year 190001, end 201812


observation = pd.DataFrame(FLOW[:,:],index = dateindex, columns = COLUMNS) # 

# 
observation_slice = observation[observation.columns.intersection(list(relating_df['gauge_id'].astype(str)))]
observation_slice = observation_slice.resample('D').ffill()

observation_slice = observation_slice.rename(columns=dictionary_rename)

print(observation_slice.head)
```

```python
numbering = ['a','b','c','d']
rivers = ['Nelson','Parana','Rhine','Mackenzie']
cases = ['_HDMA_CLM_','_HDMA_ERA5_','_HDMA_VIC_']
lake_conds = ['_no_lake','_lake']

dateindex = pd.date_range('1900-01','2018-12')
m = 0


for river in rivers:
    
    df = pd.DataFrame(index=dateindex)
    
    #
    relating_df_temp = relating_df[relating_df['riv_name']==river]
    lat = relating_df_temp['lat'].iloc[0].item()
    lat = "%.3f" % lat
    lon = relating_df_temp['lon'].iloc[0].item()
    lon = "%.3f" % lon
    
    
    for case in cases:
        
        for lake_cond in lake_conds:
            
            #
            temp = xr.open_dataset('../data/simulations/'+river+case+'IRFroutedRunoff'+lake_cond+'.nc')
            
            #
            temp = temp.to_dataframe()
            temp = temp.droplevel(-1, axis="index")
            temp.to_csv('test.csv')
            temp = pd.read_csv('test.csv')
            os.remove('test.csv')
            temp ['time'] = pd.DatetimeIndex (temp['time'])
            temp = temp.set_index(['time'])
            temp = temp.rename_axis(index=None)
            temp.index = temp.index.normalize()
            temp = temp.reindex(dateindex)
            
            df[case+'IRF'+lake_cond]=temp['IRFroutedRunoff']


    #
    df = df ['2000-01-01':'2010-01-01']
    
    fig, ax = plt.subplots(figsize=(40,10))

    observation_slice[river].plot(ax=ax, label='Observation', style= '-k',linewidth=3.0)
    df['_HDMA_CLM_IRF_no_lake'].plot(ax=ax, label='CLM no lake', style= '--b',linewidth=3.0)
    df['_HDMA_CLM_IRF_lake'].plot(ax=ax, label='CLM lake',style= '-b',linewidth=3.0)
    df['_HDMA_ERA5_IRF_no_lake'].plot(ax=ax, label='ERA5 no lake',style= '--g',linewidth=3.0)
    df['_HDMA_ERA5_IRF_lake'].plot(ax=ax, label='ERA5 lake',style= '-g',linewidth=3.0)
    df['_HDMA_VIC_IRF_no_lake'].plot(ax=ax, label='VIC no lake',style= '--r',linewidth=3.0)
    df['_HDMA_VIC_IRF_lake'].plot(ax=ax, label='VIC lake',style= '-r',linewidth=3.0)

    plt.xlim(pd.Timestamp('2000-01-01'), pd.Timestamp('2010-01-01'))
    if m ==0:
        plt.legend(fontsize=30, loc = 'upper left')
    plt.ylabel('Discharge [m³/s]',fontsize=50)
    plt.xlabel('Date',fontsize=50)
    plt.title('('+numbering[m]+') '+river+' - lon: '+str(lon)+' lat: '+str(lat),fontsize=70)
    plt.grid()
    
    
    # 
    plt.savefig('../frame/'+river+'.png')
    
    #
    m = m + 1
```
